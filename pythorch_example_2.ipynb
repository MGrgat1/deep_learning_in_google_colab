```python
import torch
import numpy as np
import math
import matplotlib.pyplot as plt

print("Torch, version:")
print(torch.__version__)
print("Cuda enabled:")
print(torch.cuda.is_available())
```

Create and plot a sine function

```python
x = np.linspace(-math.pi, math.pi, 2000)
y = np.sin(x)
plt.plot(x, y, '-r', label='y=sin(x)')
plt.title('Graph of y=sin(x)')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

Train a four-parameter polynomial to approximate a given function

```python
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()

learning_rate = 1e-6

for t in range(2001):
    # forward propagation 
    # predict the value using a polynomial with four parameters
    y_pred = a + b * x + c * x ** 2 + d * x ** 3

    # calculate the mean square error
    loss = np.square(y_pred - y).sum()
    if t % 100 == 99:
        print(t, loss)

    # error backpropagation
    # partial derivatives of a, b, c, d with respect to the error
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()

    # update parameters of the polynomials
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d

    # a graph that displays each iteration of learning
    if t % 50 == 0:
        y_graph = a + b * x + c * x ** 2 + d * x ** 3
        plt.plot(x, y_graph, '-r', label='t = ' + str(t) + ' y=a + b * x + c * x ** 2 + d * x ** 3')
        plt.title('Graph of y = a + bx + cx^2 + dx^3')
        plt.xlabel('x', color='#1C2833')
        plt.ylabel('y', color='#1C2833')
        plt.legend(loc='upper left')
        plt.grid()
        plt.show()

print(f'Result: y = {a} + {b} x + {c} x^2 + {d} x^3')
```
